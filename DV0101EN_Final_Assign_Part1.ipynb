####  TASK 1.2: Plot different lines for categories of vehicle type and analyse the trend to answer the question Is there a noticeable difference in sales trends between different vehicle types during recession periods?


<details><summary>Click here for a hint</summary>

<p>
    You will require:-
    <br>create a separate dataframe where the column recession has a value of '1'
    <br>to group the year, vehicle_type and calculate the average on the 'Automobile Sales'
    <br>one way is to -
    <br>use as_index as false else you will endup with multiple-indexed datafame
    <br>later set year as index and groupby vehicle over Sales and plot
    <br>make use of .plot() with kind = 'line'
    <br>do not forget to include labels and title
</p>
</details>


````python
import matplotlib.pyplot as plt

# Filter recession years
df_rec = df[df['Recession'] == 1]

# Avg sales by year & vehicle type during recessions
df_Mline = df_rec.groupby(['Year', 'Vehicle_Type'], as_index=False)['Automobile_Sales'].mean()

# Normalize by vehicle-type average (so lines are comparable)
df_Mline['Normalized_Sales'] = df_Mline.groupby('Vehicle_Type')['Automobile_Sales'].transform(lambda x: x / x.mean())

# Index by Year
df_Mline.set_index('Year', inplace=True)

# Plot
# vt is vehicle type
plt.figure(figsize=(12, 8))
for vt in df_Mline['Vehicle_Type'].unique():
    data = df_Mline[df_Mline['Vehicle_Type'] == vt]
    plt.plot(data.index, data['Normalized_Sales'], label=vt, marker='o')

# Mark recession years
recession_years = sorted(df_rec['Year'].unique())
for year in recession_years:
    plt.axvline(x=year, color='gray', linestyle='--', alpha=0.5)

plt.legend(title="Vehicle Type", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.ylabel("Normalized Sales")
plt.xlabel("Year")
plt.title("Normalized Automobile Sales by Vehicle Type During Recession")
plt.tight_layout()
plt.savefig('Line_Plot_2.png')
plt.show()
````

### TASK 1.5: Develop a Bubble plot for displaying the impact of seasonality on Automobile Sales.
<br>How has seasonality impacted the sales, in which months the sales were high or low? Check it for non-recession years to understand the trend

##### Develop a Bubble plot for displaying Automobile Sales for every month and use Seasonality Weight for representing the size of each bubble<br>
Title this plot as 'Seasonality impact on Automobile Sales'


<details><summary>Click here for a hint</summary>

<p>
     You can create Bubble Chart by calling the scatter()
        <br>Pass the 'Month' and 'Automobile_Sales' to the functions as x and y and then use Seasonality weight for size parameter</p>
</details>


````python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Step 1: Filter for non-recession years to understand normal seasonal trends
print("=== ANALYZING SEASONALITY DATA ===")

# Filter for non-recession periods (Recession = 0)
non_recession = df[df['Recession'] == 0].copy()
print(f"Non-recession data points: {len(non_recession)}")

# Check available data
print("Available months in data:", sorted(non_recession['Month'].unique()))
print("Date range:", non_recession['Date'].min(), "to", non_recession['Date'].max())

# Step 2: Aggregate data by month for bubble plot
# Group by month and calculate average sales and seasonality weight
monthly_data = non_recession.groupby('Month').agg({
    'Automobile_Sales': 'mean',  # Average sales per month
    'Seasonality_Weight': 'mean'  # Average seasonality weight
}).round(2)

print("\nMonthly aggregated data:")
print(monthly_data)

# Step 3: Prepare data for bubble plot
months = monthly_data.index
sales = monthly_data['Automobile_Sales']
seasonality_weights = monthly_data['Seasonality_Weight']

# Convert month names to numbers for x-axis positioning
month_mapping = {
    'Jan': 1, 'Feb': 2, 'Mar': 3, 'Apr': 4, 'May': 5, 'Jun': 6,
    'Jul': 7, 'Aug': 8, 'Sep': 9, 'Oct': 10, 'Nov': 11, 'Dec': 12
}

# Create month numbers for x-axis
month_numbers = [month_mapping.get(month, 0) for month in months]

# Step 4: Create the bubble plot
plt.figure(figsize=(14, 8))

# Create bubble plot - bubble size represents seasonality weight
# Scale bubble sizes for better visualization (multiply by large factor)
bubble_sizes = seasonality_weights * 1000  # Scale factor for visibility

# Create scatter plot with varying bubble sizes
scatter = plt.scatter(month_numbers, sales,
                     s=bubble_sizes,
                     alpha=0.6,
                     c=seasonality_weights,
                     cmap='viridis',
                     edgecolors='black',
                     linewidth=2)

# Customize the plot
plt.title('Seasonality Impact on Automobile Sales', fontsize=18, fontweight='bold', pad=20)
plt.xlabel('Month', fontsize=14, fontweight='bold')
plt.ylabel('Average Automobile Sales', fontsize=14, fontweight='bold')

# Set x-axis ticks to show month names
plt.xticks(range(1, 13), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun',
                          'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])

# Add colorbar to show seasonality weight scale
cbar = plt.colorbar(scatter)
cbar.set_label('Seasonality Weight', fontsize=12, fontweight='bold')

# Add grid for better readability
plt.grid(True, alpha=0.3)

# Add annotations for each bubble showing exact values
for i, month in enumerate(months):
    if i < len(month_numbers):  # Safety check
        plt.annotate(f'{sales.iloc[i]:.0f}',
                    (month_numbers[i], sales.iloc[i]),
                    xytext=(5, 5), textcoords='offset points',
                    fontsize=10, fontweight='bold',
                    bbox=dict(boxstyle='round,pad=0.3', facecolor='white', alpha=0.7))

# Improve layout
plt.tight_layout()

# Step 5: Analysis summary
print("\n=== SEASONALITY ANALYSIS SUMMARY ===")
max_sales_month = months[sales.argmax()]
min_sales_month = months[sales.argmin()]
max_seasonality_month = months[seasonality_weights.argmax()]
min_seasonality_month = months[seasonality_weights.argmin()]

print(f"Highest sales month: {max_sales_month} ({sales.max():.0f} sales)")
print(f"Lowest sales month: {min_sales_month} ({sales.min():.0f} sales)")
print(f"Highest seasonality impact: {max_seasonality_month} ({seasonality_weights.max():.2f})")
print(f"Lowest seasonality impact: {min_seasonality_month} ({seasonality_weights.min():.2f})")

# Calculate seasonal variation
sales_range = sales.max() - sales.min()
seasonality_range = seasonality_weights.max() - seasonality_weights.min()
print(f"\nSales variation: {sales_range:.0f} units ({(sales_range/sales.mean()*100):.1f}% of average)")
print(f"Seasonality variation: {seasonality_range:.2f} ({(seasonality_range/seasonality_weights.mean()*100):.1f}% of average)")

#Save the figure
plt.savefig('Bubble.png')
# Show the plot
plt.show()

# Step 6: Additional insights table
print("\n=== DETAILED MONTHLY BREAKDOWN ===")
analysis_df = pd.DataFrame({
    'Month': months,
    'Avg_Sales': sales.round(0),
    'Seasonality_Weight': seasonality_weights.round(3),
    'Sales_Rank': sales.rank(ascending=False).astype(int),
    'Seasonality_Rank': seasonality_weights.rank(ascending=False).astype(int)
})

print(analysis_df.to_string(index=False))
````
